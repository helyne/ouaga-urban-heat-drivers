# GCCM write up sections

## Methods

To distinguish causal drivers of LST from mere correlates, we applied Geographical Convergent Cross Mapping (GCCM; Gao et al., 2023) as implemented in the spEDM R package (Lyu, 2026). GCCM extends Sugihara et al.'s (2012) convergent cross mapping to spatial cross-sectional data by reconstructing attractor manifolds from spatially lagged embeddings and testing whether one variable's state space can predict another's — a signature of causal influence under Takens' theorem. For each predictor–LST pair, we ran bidirectional GCCM on the 150 m aggregated raster with embedding dimension E = 3 and spatial lag τ = 1 (150 m), using 2,000 randomly sampled prediction points. Causal direction was determined by three criteria following Gao et al.: (1) convergence (Kendall's τ > 0 for cross-map skill ρ vs. library size), (2) significance at the largest library size (p < 0.05, two-tailed t-test on ρ with nominal N = 2,000), and (3) non-overlapping 95% Fisher-z confidence intervals between forward and reverse directions. We report Clifford/Dutilleul effective sample sizes (N_eff = 54–662) as supplementary context for the degree of spatial autocorrelation. Sensitivity analyses with alternative spatial lags (τ = 5, 10) and per-variable simplex E-selection confirmed that causal directions were robust to these choices.

## Results

Causal validation of predictive features. To distinguish genuinely causal drivers from correlational predictors identified by the ML model, we applied GCCM to each predictor–LST pair. Built-up density (ρ = 0.49, 95% CI [0.46, 0.53]) and green density (ρ = 0.34 [0.30, 0.38]) showed clear unidirectional causality toward LST, with reverse cross-map skill substantially weaker (ρ = 0.21 and 0.16, respectively) and non-overlapping confidence intervals. Distance to roads also showed asymmetric coupling favouring a causal effect on LST (ρ = 0.25 vs. 0.18), though with overlapping CIs. In contrast, the spectral indices that ranked highly in the ML model (NDBI, BSI, NDVI; ρ = 0.50–0.73) exhibited symmetric bidirectional coupling, indicating strong association but no identifiable causal direction — consistent with shared radiometric properties between these indices and surface temperature. Distance to water similarly showed bidirectional coupling (ρ ≈ 0.40–0.45 both ways). All predictors exhibited robust convergence (Kendall's τ ≥ 0.94, p < 10⁻⁶). These results suggest that while spectral indices are effective predictors of LST, the actionable causal levers for urban heat mitigation in Ouagadougou are the physical land surface properties: built-up density, green cover, and road infrastructure.

## Discussion

Causal interpretation. The GCCM analysis reveals an important divergence between predictive importance and causal influence. Built-up density was both the most predictive feature (mean |SHAP| ≈ 2.2) and the strongest confirmed causal driver of LST (ρ = 0.49 forward vs. 0.21 reverse, non-overlapping 95% CIs). Green vegetation density, by contrast, ranked sixth in SHAP importance but was the second variable for which GCCM confirmed unidirectional causality toward LST (ρ = 0.34 vs. 0.16). Its low predictive ranking likely reflects redundancy with built-up density and NDVI in the ML model rather than a weak causal role — a distinction that purely predictive analyses would miss. 

The spectral indices NDBI, BSI, and NDVI ranked highly in SHAP (4th, 5th, and 7th) but showed only symmetric bidirectional coupling in GCCM (ρ = 0.50–0.73 in both directions), precluding directional inference. This is physically expected: these indices share radiometric properties with thermal emission and are derived from overlapping spectral information, making them strong statistical proxies for LST without being upstream causes of it. Similarly, DEM — the third most important predictor — exhibited near-equal cross-mapping skill in both directions (ρ ≈ 0.44), likely reflecting the limited topographic relief across Ouagadougou (~40 m range), which provides insufficient variation for GCCM to resolve directionality. 

These results have direct implications for urban heat mitigation. While spectral indices and elevation are useful for predicting where hotspots occur, they are not intervention targets. The features that GCCM confirmed as causal — built-up density and green cover — correspond to actionable urban planning levers: reducing impervious surface fraction and increasing vegetation. 


Notably, Ouagadougou's hotspot pattern inverts the classical urban heat island. The city centre, which retains more mature vegetation and water features, is relatively cool, consistent with the nocturnal cool island documented by Lindén (2011). The rapidly expanding peri-urban outskirts, dominated by bare soil, impervious surfaces, and sparse vegetation, are the hottest areas. This suggests that heat mitigation efforts should prioritise greening and reducing bare surface fraction in these peripheral zones, where urbanisation is outpacing the establishment of tree cover and green infrastructure. The finding aligns with Oke et al.'s (2017) broader framework linking surface energy balance to land cover composition, though it underscores that in semi-arid Sahelian cities, the standard "cool suburbs, hot core" assumption does not hold.
